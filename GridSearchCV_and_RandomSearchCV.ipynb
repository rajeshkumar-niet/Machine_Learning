```python
# Import libraries
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

## Data Preprocessing

```python
dataset = sklearn.datasets.load_breast_cancer()
dataset.keys()
print(dataset)
```

```python
dataframe = pd.DataFrame(dataset.data, columns=dataset.feature_names)
dataframe['label'] = dataset.target
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

```python
dataframe['label'].value_counts()
```

```python
X = dataframe.drop(columns='label', axis=1)
Y = dataframe['label']
```

```python
print(X.shape, Y.shape)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

## Grid Search CV
Best hypermeter tuning

```python
from sklearn.svm import SVC
model = SVC()
```

```python
# Hyperparameter

parametters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1, 5, 10, 20]
}
```

```python
gridclassifier = GridSearchCV(estimator=model, param_grid=parametters, cv=5)
gridclassifier.fit(X, Y)
```

```python
gridclassifier.cv_results_
```

```python
gridclassifier.best_params_
```

```python
highest_accuracy = gridclassifier.best_score_
print(round(highest_accuracy*100, 2))
```

```python
gird_result = pd.DataFrame(gridclassifier.cv_results_)
gird_result
```

```python
gird_result_main = gird_result[['param_C', 'param_kernel','mean_test_score']]
gird_result_main
```

```python

gird_result_main.to_csv('grid_search_results.csv', index=False)
```

## Randomized Grid Search CV


```python
X = dataframe.drop(columns='label', axis=1)
Y = dataframe['label']
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
parametters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1, 5, 10, 20]
}
```

```python
random_classifier = RandomizedSearchCV(estimator=model, param_distributions=parametters, cv=5)
```

```python
random_classifier.fit(X, Y)
```

```python
random_classifier.cv_results_
```

```python
random_classifier.best_params_
```

```python
random_result = pd.DataFrame(random_classifier.cv_results_)
random_result
```

```python
random_result_main = random_result[['param_C', 'param_kernel','mean_test_score']]
```

```python
random_result_main
```

```python
random_result_main.to_csv('randomized_search_results.csv', index=False)
```
