### Dot product of two vectors
 product is always a scaler

```python
import numpy as np
```

```python
a = np.array([2,3])
b = np.array([4,5])
# numpy funtion to find dot product is dot()
a_dot_b = np.dot(a,b)
print("Dot product of a and b : ", a_dot_b)
```

```python
x = np.array([5,7,3])
y = np.array([9,4,2])
x_dot_y = np.dot(x,y)
print("Dot product of x and y : ",x_dot_y)
```

### Cross Product

Product is vector

```python
a = np.array([2,3])
b = np.array([4,5])
# numpy funtion to find cross product is cross()
a_cross_b = np.cross(a,b)
print("Cross product of a and b : ",a_cross_b)
```

```python
x = np.array([5,7,3])
y = np.array([9,4,2])
x_cross_y = np.cross(x,y)
print("Cross product of x and y : ", x_cross_y)
```

### Projection of A on B vector

```python
a = np.array([2,3])
b = np.array([4,6])

# Magnitude of b
mb = np.sqrt(sum(b**2))

projection_of_a_on_b = (np.dot(a,b)/mb)*b
print("Projection of a on b : ",projection_of_a_on_b)
```

```python
a = np.array([4,7,9])
b = np.array([9,4,8])

# Magnitude of b
mb = np.sqrt(sum(b**2))

projection_of_a_on_b = (np.dot(a,b)/mb)*b
print("Projection of a on b : ",projection_of_a_on_b)
```

```python
x = np.random.randint(1,100,(4,5))
y = np.random.randint(1,100,(4,5))
```

```python
# Multiplication of two matrixes
try:
  x = np.random.randint(1,10,(4,3))
  y = np.random.randint(1,10,(3,5))
  product = np.dot(x,y)
  print(product)
except:
  print("Multiplication not possible ")
```
