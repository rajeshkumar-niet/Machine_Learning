```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# Importing models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
X = dataset.drop(columns='target', axis=1)
Y = dataset['target']
```

```python
print(X.shape)
print(Y.shape)
```

```python
X = np.array(X)
Y = np.array(Y)
```

### Model Selection
Comparing all models and selecting best of these

```python
models = [LogisticRegression(max_iter=1000), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier(), SVC(kernel='linear')]
```

```python
def compare_models():
    for model in models:
        cv_score = cross_val_score(model,X,Y, cv=5)
        cv_score = np.array(cv_score)
        mean_accuracy = sum(cv_score)/len(cv_score)
        mean_accuracy = mean_accuracy*100
        mean_accuracy = round(mean_accuracy, 2)
        print('Cross Validation accuracies for the',model,'=', cv_score)
        print('Accuracy score of the ',model,'=',mean_accuracy,'%')
        print('------------------------------------------------------')
```

```python
compare_models()
```

```python
models = [LogisticRegression(max_iter=1000), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
model_hyperparameters = {
    'log_reg_hyperparameters': {
        'C' : [1,5,10,20]
    },
    'decision_tree_hyperparameters' : {
        'criterion' : ['gini', 'entropy']
    },
    'random_forest_hyperparameters' : {
        'n_estimators' : [10, 20,30]
        },
    'knn_hyperparameters' : {
        'n_neighbors' : [3,5,10]
        },
}
```

```python
model_keys = list(model_hyperparameters.keys())
print(model_keys)
```

###GridSearchCV

```python
def model_selection_grid(list_of_models, hyperparameters_dictionary):
    result = []
    i = 0
    for model in list_of_models:
        keys = hyperparameters_dictionary[i]
        params = model_hyperparameters[keys]
        grid_search = GridSearchCV(model, params, cv=5)
        grid_search.fit(X,Y)
        i+=1
        print("Model : ", model)
        print("Params : ", params)
        print("Best Score : ", round(grid_search.best_score_*100, 2))
        print( 'best hyperparameters' , grid_search.best_params_)
        print("------------------------------------------------------")
        result.append({
            'model used' : model,
            'highest score' : round(grid_search.best_score_*100, 2),
            'best hyperparameters' : grid_search.best_params_
        })

    result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])

    return result_dataframe
```

```python
result = model_selection_grid(models, model_keys)
```
