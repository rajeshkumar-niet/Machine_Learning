# MatplotLib


*   Usefull for making plots


```python
# Impoting libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
# Sine Wave
plt.plot(x,y,"p")
plt.xlabel("Angle")
plt.ylabel("Sin")
plt.title("My First plot SIN(X) Plot")
plt.show()
```

```python
# Cos Wave
plt.plot(x,z)
plt.xlabel("Angle")
plt.ylabel("Cos")
plt.title("My First plot COS(X) Plot")
```

```python
# Parabola
x = np.linspace(-10,10,50)
y = x**2
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Parabola")
plt.show()
```

```python
# Two graphs in a same plot
x = np.linspace(0,10,50)
plt.plot(x, np.sin(x),"r")
plt.plot(x, np.cos(x),"g")
plt.xlabel("Angel in radian")
plt.ylabel("Sin And Cos ")
plt.title("Cos and Sin in a graph")
```

## Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
lang = ["Hindi", "Eng", "Urdu"]
public = [100,150,75]
ax.bar(lang, public)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.21,0.21])
x = np.linspace(1,10,11)
ax.bar(x , np.sin(x))
plt.show()
```

## Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
lang = ["Hindi", "Eng", "Urdu", "Tamil","Kannad"]
public = [100,95,75, 50,35]
ax.pie(public, labels= lang, autopct ="%1.2f%%")
plt.title("Langues")
```

## Scatter Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.linspace(0,10,50)
ax.scatter(x,np.sin(x))
ax.scatter(x,np.cos(x))
plt.show()
```

## 3D Scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = 10* np.random.random(10)
y = np.sin(x)
z = np.cos(x)
ax.scatter(z,y,x, c = x, cmap='Blues')
plt.show()
```
